```python
import cv2
```

```python
img = cv2.imread('img/lion.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow('LION',img)
cv2.waitKey()
cv2.destroyWindow('LION')
```

```python
equalised_img = cv2.equalizeHist(img)
```

```python
cv2.imshow('EQ. IMG',equalised_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
plt.imshow(equalised_img)
```

```python
hist_arr = cv2.calcHist([img],[0],None,[256],[0,256])
```
